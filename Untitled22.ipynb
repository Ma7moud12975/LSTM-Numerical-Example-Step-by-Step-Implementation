```python
import math
from dataclasses import dataclass
from typing import List, Dict

# -----------------------------
# Helper math functions
# -----------------------------
def sigmoid(x: float) -> float:
    return 1 / (1 + math.exp(-x))

def tanh(x: float) -> float:
    return math.tanh(x)

def line(char="=", width=70):
    print(char * width)

def section(title: str):
    print("\n")
    line("=")
    print(f" {title}")
    line("=")

def show_formula(label: str, expr: str, value: float, decimals: int = 3):
    print(f"{label:<18} {expr} = {value:.{decimals}f}")
```

```python
@dataclass
class LSTMParams:
    Wf: float; Whf: float; bf: float
    Wi: float; Whi: float; bi: float
    Wc: float; Whc: float; bc: float
    Wo: float; Who: float; bo: float

@dataclass
class OutputLayer:
    Wy: float; by: float

@dataclass
class LSTMState:
    h: float = 0.0
    c: float = 0.0
```

```python
def lstm_step(x_t: float, prev_state: LSTMState, params: LSTMParams) -> Dict[str, float]:
    h_prev = prev_state.h
    c_prev = prev_state.c

    # Gates
    f_t_input = params.Wf * x_t + params.Whf * h_prev + params.bf
    i_t_input = params.Wi * x_t + params.Whi * h_prev + params.bi
    c_hat_input = params.Wc * x_t + params.Whc * h_prev + params.bc
    o_t_input = params.Wo * x_t + params.Who * h_prev + params.bo

    f_t = sigmoid(f_t_input)
    i_t = sigmoid(i_t_input)
    c_hat_t = tanh(c_hat_input)
    c_t = f_t * c_prev + i_t * c_hat_t
    o_t = sigmoid(o_t_input)
    h_t = o_t * tanh(c_t)

    return {
        "x_t": x_t, "h_prev": h_prev, "c_prev": c_prev, "f_t": f_t,
        "i_t": i_t, "c_hat_t": c_hat_t, "c_t": c_t, "o_t": o_t, "h_t": h_t
    }
```

```python
def print_time_step(t: int, result: Dict[str, float]):
    x_t = result["x_t"]; h_prev = result["h_prev"]; c_prev = result["c_prev"]
    print(f"\n🟣 Time Step t = {t}, Input x_{t} = {x_t}")
    line("-")
    show_formula("1) Forget gate:", f"σ(0.5*{x_t} + 0.1*{h_prev:.3f} + 0)", result["f_t"])
    show_formula("2) Input gate:", f"σ(0.6*{x_t} + 0.2*{h_prev:.3f} + 0)", result["i_t"])
    show_formula("3) Candidate:", f"tanh(0.7*{x_t} + 0.3*{h_prev:.3f} + 0)", result["c_hat_t"])
    c_new_from_old = result["f_t"] * c_prev
    c_new_from_candidate = result["i_t"] * result["c_hat_t"]
    print(f"4) Cell state:      ({result['f_t']:.3f} * {c_prev:.3f}) + ({result['i_t']:.3f} * {result['c_hat_t']:.3f}) = {result['c_t']:.3f}")
    show_formula("5) Output gate:", f"σ(0.8*{x_t} + 0.4*{h_prev:.3f} + 0)", result["o_t"])
    print(f"6) Hidden state:    {result['o_t']:.3f} * tanh({result['c_t']:.3f}) = {result['h_t']:.3f}")

def run_lstm_example():
    section("LSTM NUMERICAL EXAMPLE")
    X = [1, 2, 3]
    params = LSTMParams(Wf=0.5, Whf=0.1, bf=0.0, Wi=0.6, Whi=0.2, bi=0.0, Wc=0.7, Whc=0.3, bc=0.0, Wo=0.8, Who=0.4, bo=0.0)
    output_layer = OutputLayer(Wy=4.0, by=0.0)
    state = LSTMState(h=0.0, c=0.0)

    history = []
    for t, x_t in enumerate(X, start=1):
        result = lstm_step(x_t, state, params)
        print_time_step(t, result)
        state = LSTMState(h=result["h_t"], c=result["c_t"])
        history.append(result)

    section("STEP 2 - PREDICT THE NEXT VALUE")
    result_t4 = lstm_step(4, state, params)
    print_time_step(4, result_t4)
    y_hat = output_layer.Wy * result_t4["h_t"] + output_layer.by
    print(f"\nFinal Prediction: {y_hat:.3f}")

if __name__ == "__main__":
    run_lstm_example()
```
